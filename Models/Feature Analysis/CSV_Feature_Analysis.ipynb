```python
# features with no variance (should be none but just in case)
no_var_inop = feature_var_inoptimal[feature_var_inoptimal == 0].index
no_var_op = feature_var_optimal[feature_var_optimal == 0].index

# features with variance below threshold
var_thres_inop = feature_var_inoptimal[feature_var_inoptimal < 0.01].index
var_thres_op = feature_var_optimal[feature_var_optimal < 0.01].index
```

```python
print("aInoptimal \n")
print("Features with No Variance:", list(no_var_inop))
print("Features with Low Variance:", list(var_thres_inop))

print("\naOptimal \n")
print("Features with No Variance:", list(no_var_op))
print("Features with Low Variance:", list(var_thres_op))
```

# High Correlation Features

```python
correlation_matrix_inop = aInoptimal.corr()
correlation_matrix_op = aOptimal.corr()
```

```python
# identify highly correlated features


# for Inoptimal data
high_correlation_pairs_inop = correlation_matrix_inop.abs().stack().reset_index()

# .stack produces new columns renamted to the below, when feature1 == feature2 correlation is counted
high_correlation_pairs_inop.columns = ['Feature1', 'Feature2', 'Correlation']  
high_correlation_pairs_inop = high_correlation_pairs_inop[high_correlation_pairs_inop['Feature1'] != high_correlation_pairs_inop['Feature2']]
high_correlation_pairs_inop = high_correlation_pairs_inop[high_correlation_pairs_inop['Correlation'] > 0.9]

# for Optimal data
high_correlation_pairs_op = correlation_matrix_op.abs().stack().reset_index()
high_correlation_pairs_op.columns = ['Feature1', 'Feature2', 'Correlation']  
high_correlation_pairs_op = high_correlation_pairs_op[high_correlation_pairs_op['Feature1'] != high_correlation_pairs_op['Feature2']]
high_correlation_pairs_op = high_correlation_pairs_op[high_correlation_pairs_op['Correlation'] > 0.9]
```

### No 90+% correlation for inoptimal but slider can be adjusted depending on our needs

```python
print("Highly Correlated Feature Pairs (Inoptimal):\n", high_correlation_pairs_inop)
print("\nHighly Correlated Feature Pairs (Optimal):\n", high_correlation_pairs_op)
```
